# ***<font color="RoyalBlue">Download FASTQ (fasterq-dump)</font>***

Download FASTQ files from SRA (Sequence Read Archive) and convert them to simple filenames.

## 1. Create Accession Number List and Run Download

Create a list of Accession numbers (e.g., DRR147067) of the data you want to download, and execute the download in one step.

- Enter the Accession number list after "user_input = ("""" (one number per line).
- You can adjust the number of threads (`threads`) if necessary.

<font color="Tomato">────────────── 	↓↓↓ 	***Run Command***	 ↓↓↓	 ──────────────</font>

```python
####################################################
# Set Parameters
####################################################
# List file name (No need to change)
file_name = "list.txt"

# Paste the Accession number list
user_input = ("""
DRR147067
DRR147068
DRR147069
"""
)

# Number of threads
threads = 8

####################################################
# *** Do not change below this line ***
####################################################

# 1. Create the list
# Remove leading/trailing newlines and add a trailing newline
user_input = user_input.strip("\n") + "\n"

# Write the input to the file
print(f"1. Creating Accession Number List")
with open(file_name, "w") as file:
    file.write(user_input)
print(f"{file_name} has been created.")
print("")
print("-" * 30)

print(f"2. Downloading FASTQ files")
# Output directory
output_dir = "."
# 2. Run fasterq-dump
!bash fasterq_dump.sh \
    -i $file_name \
    -t $threads \
    -o $output_dir

!echo '\nComplete!'
```
