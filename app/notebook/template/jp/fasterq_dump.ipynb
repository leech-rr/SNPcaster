# ***<font color="RoyalBlue">FASTQダウンロード (fasterq-dump)</font>***

SRA (Sequence Read Archive) からFASTQファイルをダウンロードし、シンプルなファイル名に変換します。

## 1. Accession numberリストの作成とダウンロード

ダウンロードしたいデータのAccession number (例: DRR147067) のリストを作成し、そのままダウンロードを実行します。

- 「user_input = ("""」の後に、Accession numberリストを入力してください（1行に1つの番号）。
- スレッド数(`threads`)は用途に合わせて変更してください。

<font color="Tomato">────────────── 	↓↓↓ 	***コマンド実行***	 ↓↓↓	 ──────────────</font>

```python
####################################################
# パラメータ設定
####################################################
# リストの出力ファイル名 (変更不要)
file_name = "list.txt"

# Accession numberリストを貼り付ける
user_input = ("""
DRR147067
DRR147068
DRR147069
"""
)

# スレッド数
threads = 8

####################################################
# ※以下は変更しないでください※
####################################################

# 1. リストの作成
# 前後の改行を除去後に改行追加
user_input = user_input.strip("\n") + "\n"

# 入力された要素をファイルに書き込む
print(f"1.アクセッションナンバーリストの作成")
with open(file_name, "w") as file:
    file.write(user_input)
print(f"{file_name} を作成しました。")
print("")
print("-" * 30)

print(f"2.FASTQファイルのダウンロード")
# 出力ディレクトリ
output_dir = "."
# 2. fasterq-dumpの実行
!bash fasterq_dump.sh \
    -i $file_name \
    -t $threads \
    -o $output_dir

!echo '\nComplete!'
```
