# ***<u><font color="RoyalBlue">解析手順</font></u>***


## 1. 入力ファイルのアップロード

解析プロジェクトのフォルダへ、FASTAファイル（アセンブリ済みのファイル）をアップロードしてください。  
左サイドバー上部の⬆︎ボタンから、または、解析プロジェクトフォルダを開きファイルをドラック＆ドロップでアップロードできます。
<br>


## 2. 菌株リストの作成

次のいずれかの方法で、菌株リストを作成します。

### 2.1. Jupyter Labで作成する
- 左サイドバー上部の「 + 」ボタン→「Other」の「Text file」を押下し、ファイルを新規作成できる。
- 編集が完了したら、Ctrl＋s（Mac:Command+s）を押下し、ファイル名を設定して保存する。
### 2.2.  Notepad、メモ帳、Vim等(何でも良い)で作成して、解析フォルダにコピーする。
- Windowsで作成すると、正常に動かない可能性があります。（改行コードの違いのため）
- LinuxまたはMacで作成してください。
### 2.3. 下記コマンドで作成する。       
- 「file_name = 」の後に、リスト名をダブルクオート(")で囲って入力してください（例, "ファイル名") 
- 「user_input = ("""」の後に、菌株リストを入力してください（1行に1つの株名）。

<font color="Tomato">────────────── 　↓↓↓ 　***コマンド実行***　 ↓↓↓　 ──────────────</font>

```python
# リストのファイル名を入力
file_name = "list_amrfinder.txt"

# 菌株リストを貼り付ける
user_input = ("""
A0001
A0002
A0003
A0004
A0005
A0006
A0007
A0008
"""
)

# 前後の改行を除去後に改行追加
user_input = user_input.strip("\n") + "\n"

# 入力された要素をファイルに書き込む
with open(file_name, "w") as file:
    file.write(user_input)
!echo 'Complete!'
```

<font color="Tomato">────────────────────────────────────────────</font>
<br>

## 3. amrfinder.shの実行

- 「file_name = 」の後に、前の項で作成した株名リスト名をダブルクオート(")で囲って入力してください。 
- `パラメータを設定`に記載されている内容で修正したいものがあれば修正し、実行してください。
    - 記載されている内容のまま実行することも可能です。

<font color="Tomato">────────────── 　↓↓↓ 　***コマンド実行***　 ↓↓↓　 ──────────────</font>

```python
import subprocess

####################################################
# パラメータを設定
####################################################
# 2.で作成した菌株リストのファイル名を入力
file_name = "list_amrfinder.txt"
# FASTAファイルの拡張子を入力
file_extension = ".fasta"
# スレッド数を入力
threads = "8"
# plus機能(耐性遺伝子に加えて、病原性遺伝子や金属耐性遺伝子等を検出する)を
# 使用する場合は"1"、使用しない場合は"0"を入力
plus = "1"
# 属/種の指定を行う場合、
# 以下の一覧から1つのみ行頭の# を外して有効化してください。
# 属/種が不明な場合や一覧にない場合は、すべて# を付けたままにしてください(精度は下がります)。
# species = "Acinetobacter_baumannii"
# species = "Bordetella_pertussis"
# species = "Burkholderia_cepacia"
# species = "Burkholderia_mallei"
# species = "Burkholderia_pseudomallei"
# species = "Campylobacter"
# species = "Citrobacter_freundii"
# species = "Clostridioides_difficile"
# species = "Corynebacterium_diphtheriae"
# species = "Enterobacter_asburiae"
# species = "Enterobacter_cloacae"
# species = "Enterococcus_faecalis"
# species = "Enterococcus_faecium"
# species = "Escherichia"
# species = "Haemophilus_influenzae"
# species = "Helicobacter_pylori"
# species = "Klebsiella_oxytoca"
# species = "Klebsiella_pneumoniae"
# species = "Neisseria_gonorrhoeae"
# species = "Neisseria_meningitidis"
# species = "Pseudomonas_aeruginosa"
# species = "Salmonella"
# species = "Serratia_marcescens"
# species = "Staphylococcus_aureus"
# species = "Staphylococcus_pseudintermedius"
# species = "Streptococcus_agalactiae"
# species = "Streptococcus_pneumoniae"
# species = "Streptococcus_pyogenes"
# species = "Vibrio_cholerae"
# species = "Vibrio_parahaemolyticus"
# species = "Vibrio_vulnificus"


####################################################
# Run amrfinder_batch.sh
# ※以下は変更しないでください※
####################################################
# Run amrfinder_batch.sh
plus_option = f"--plus " if plus == "1" else ""
species_option = f"-s {species} " if 'species' in locals() else ""

!bash amrfinder_batch.sh \
    -i $file_name \
    -e $file_extension \
    -t $threads $plus_option$species_option\
    | tee amrfinder_batch.log
!echo 'Complete!'
```

### 4.	実行後、`amrfinder_[日付]_[時間]_[リスト名]`のフォルダが作製されます。フォルダ内のファイルは次の通りです<br>

- [株名]_(plus\_)out.tsv：結果ファイルです。"plus"オプションを付加した場合、`_plus_out`となります。
- logフォルダ：ログはこのフォルダにまとめられます。解析がうまくいかないときなどにご参照ください。

## 5. 出力形式の説明

実行後に生成される結果ファイル（TSV形式）の列の説明は以下の通りです。

| 列名 | 説明 |
|------|------|
| **Protein id** | タンパク質またはDNA配列のFASTAヘッダ行から取得されるID |
| **Contig id** | コンティグ名 |
| **Start** | DNAコンティグ上の遺伝子開始位置（1ベース） |
| **Stop** | DNAコンティグ上の遺伝子終了位置（1ベース） |
| **Strand** | 配列の方向（"+"=正鎖、"-"=逆鎖） |
| **Element symbol** | 遺伝子記号またはSNP定義。点突然変異の場合は遺伝子記号と突然変異定義が"_"で結合 |
| **Element name** | タンパク質、RNA、または点突然変異の完全名 |
| **Scope** | データベースのスコープ："core"（耐性効果が期待される）または"plus"（追加の興味ある遺伝子） |
| **Type** | 機能カテゴリ："AMR"（薬剤耐性）、"STRESS"（ストレス応答）、または"VIRULENCE"（病原性） |
| **Subtype** | タイプのさらなる詳細分類 |
| **Class** | AMR遺伝子の場合、この遺伝子が関与する薬剤クラス |
| **Subclass** | 薬剤クラス内の詳細化 |
| **Method** | ヒットのタイプ（ALLELE, EXACT, BLAST, PARTIAL, HMM, INTERNAL_STOP, POINT等） |
| **Target length** | クエリタンパク質または遺伝子の長さ（アミノ酸または塩基数） |
| **Reference sequence length** | データベース内の参照配列の長さ（BLASTヒットが検出された場合） |
| **% Coverage of reference** | BLASTヒットによってカバーされた参照の割合（%） |
| **% Identity to reference** | 参照タンパク質またはヌクレオチド配列への相同性（%） |
| **Alignment length** | BLASTアラインメントの長さ（アミノ酸またはヌクレオチド） |
| **Closest reference accession** | BLASTヒットのRefSeqアクセッション番号 |
| **Closest reference name** | BLASTヒットの参照配列の完全名 |
| **HMM accession** | HMMアクセッション番号（HMM検索の場合）、それ以外の場合はNA |
| **HMM description** | HMMに関連付けられたファミリー名（HMM検索の場合）、それ以外の場合はNA |
