# ***Create Sample Project***

### This notebook creates a sample project using STEC sample data
This notebook automatically downloads STEC sample data (https://github.com/leech-rr/WGS_benchmark_dataset) and creates a sample project for analysis.<br>
The sample data consists of 3 outbreak cases (5 strains each, 15 strains total). Use this for evaluating the analysis program.

If you want to create a project using your own data, use [CreateProject_en.ipynb](CreateProject_en.ipynb).

### 1. Enter the analysis project name as in the example
- Example: projectdir = "stec_sample_project" <br>
        → A folder named "stec_sample_project" will be created **within the project folder**
- Do not include spaces
- Do not use existing folder names
- It works fine if you leave it as "stec_sample_project".

```python
project_name = "stec_sample_project"
```

### 2. Execute RUN ALL (▶▶) at the top of this notebook
The project folder specified in the previous section will be created **within the project folder**, template files will be copied, and sample data will be downloaded.<br>
Since the sample data size is large, downloading will take some time.

```python
# Number of threads to use for download
threads = 4

###################################
### Do not modify the following ###
###################################
import shutil

list = f'sample_data/stec/sample_list'
project_dir = f'project/{project_name}'
shutil.copytree(f'template/en', project_dir)
shutil.copytree(f'sample_data/stec', project_dir, dirs_exist_ok=True)

!bash fasterq_dump.sh \
    -i $list \
    -t $threads \
    -o $project_dir
!echo 'Complete!'
```
