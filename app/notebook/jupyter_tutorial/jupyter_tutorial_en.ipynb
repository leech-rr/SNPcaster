## Basic Structure of Jupyter
1. Notebook

    * Programs, execution results, graphs, work notes, and related documents are managed together in a file format called a notebook.

        <img src="contents/en/start_notebook.png" width="70%">

<br>

2. Kernel

    * A kernel is the core part that executes Python and other programs. It stores object contents in memory and performs calculations.
    Simply closing the browser window does not terminate the kernel.

<br>

3. Cell

    * In Jupyter, you write source code and explanations in parts called cells within the notebook.
    The square areas in the image below are cells.

        <img src="contents/jp/new_cell.png" width="70%">

<br>

4. Cell Execution Types

    * Cells come in different types depending on their execution format. The execution format can be selected from the list below: "Code", "Markdown", "Raw". Each has the following roles:
    
        - Code
            - Select this when you want to write and execute code in programming languages such as Python or R, or OS commands.
            After entering the program, press [Shift]+[Enter] to execute it and the results will be displayed.
        - Markdown
            - Select this when you want to save content as a document in Markdown format. You can use it as notes or to save and share explanations and supplementary information about programs along with the programs.
        - Raw
            - Can be used as plain text but is rarely used.

<br>

5. Cell Structure

    * A Markdown cell is placed above the Code cell to provide explanation for the analysis.
    After code execution, the execution results are displayed below the code cell.

    
        <img src="contents/en/Cells.png" width="70%">

## Editing Notebooks
1. Inserting a Cell

    * Click the "+" button at the top of the notebook, and a new cell will be added below the currently selected cell.<br>
    * Select the new cell and change the execution format as desired. There are 3 types: "Code", "Markdown", "Raw". The default is "Code" cell.

<img src="contents/en/add_cell.png" width="70%">

2. Executing a Cell
    * Select the cell you want to execute and click the "▶️" (Run Cell) button at the top of the notebook to execute the commands in that cell.


3. Checking Execution Status

    * If execution stops due to an error, the bracket [ ] on the left side of the cell will be empty.
    * If execution fails, check the boundary between [*] and [ ].
    * If an error occurs, fix the Code cell and execute it again using "2. Executing a Cell".
    * Also, you can check the progress by clicking the pie chart icon in the upper right.
    * The Chrome tab icon also changes to an hourglass mark during execution.

## Starting a New Analysis

1. Creating a New Analysis Folder

    * From the left menu bar, click and open "CreateProject_en.ipynb". This notebook creates your analysis working folder.
    * Edit `projectdir` in the Code cell, and after completing the input, run the entire notebook. (Click the **▶️▶️ mark** at the top)

    * After execution is complete, a new working folder with the name specified in `projectdir` will be created.


    * <img src="contents/en/click_StartAnalysis.png" width="70%">

    * Click and open the new analysis folder created in "1. Creating a New Analysis Folder".
